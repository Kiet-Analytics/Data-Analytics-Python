```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_df.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df.head()
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
df_sum = df[years].sum().to_frame(name='Total')
df_sum.reset_index(names='Years', inplace=True)
```

```python
df_sum
```

```python
xticks = np.arange(0, 34, step=3)
labels = list(map(str,np.arange(1980, 2014, 3)))

ax = df_sum.plot(kind='scatter', x='Years', y='Total', figsize=(13,8))
ax.set_xticks(xticks)
ax.set_xticklabels(labels)

plt.title('Total Immigrants Population to Canada from 1980-2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.show()

```
