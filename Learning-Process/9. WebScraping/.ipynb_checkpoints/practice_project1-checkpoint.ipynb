### Exercise 1
Extract the required GDP data from the given URL using Web Scraping.


```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
# Extract tables from webpage using Pandas. Retain table number 3 as the required dataframe.
tables = pd.read_html(URL)

df = tables[3]
df.head()
```

```python
# Replace the column headers with column numbers
df.columns = range(df.shape[1])

# Retain columns with index 0 and 2 (name of country and value of GDP quoted by IMF)
df = df[[0,2]]

```

```python
# Assign column names as "Country" and "GDP (Million USD)"
df.rename(columns = {0:'Country', 2:'GDP'}, inplace=True)
df
```

```python
# Replace '-' with 0 to change datatype for column GDP
df['GDP'] = df['GDP'].replace('-', 0, inplace=True)
df
```

```python
# change data type in the column to numberic sorting values by GDP
df['GDP'] = df['GDP'].astype('int32')
#method2
df = df.astype({'GDP': 'int32'})
            
```

```python
# Retain the Rows with index 1 to 10, indicating the top 10 economies of the world.
df = df.sort_values(by=['GDP'], ascending=False).head(10)
df
```
