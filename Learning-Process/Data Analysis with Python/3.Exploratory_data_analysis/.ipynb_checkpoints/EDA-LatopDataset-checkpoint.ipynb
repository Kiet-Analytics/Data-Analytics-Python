Generate regression plots for each of the parameters "CPU_frequency", "Screen_Size_inch" and "Weight_pounds" against "Price". Also, print the value of correlation of each feature with "Price".

# Task 3 - GroupBy and Pivot Tables

Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.


````python
# Write your code below and press Shift+Enter to execute
# Create the group
df_test = df[['GPU','CPU_core','Price']]
df_gr = df_test.groupby(['GPU', 'CPU_core'], as_index=False).mean()
df_gr
````

````python
# Create the Pivot table
df_pivot = df_gr.pivot(index='GPU', columns='CPU_core')
df_pivot
````

````python
# Create the heatmap from pivot
fig, ax = plt.subplots()
im = ax.pcolor(df_pivot, cmap='RdBu')


#set xticks and yticks
ax.set_xticks(np.arange(df_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df_pivot.shape[0]) + 0.5, minor=False)

#set labels for these thicks
y_ticklabels = df_pivot.index
x_ticklabels = df_pivot.columns.levels[1]

ax.set_xticklabels(x_ticklabels, minor=False)
ax.set_yticklabels(y_ticklabels, minor=False)

plt.ylabel('GPU')
plt.xlabel('CPU_core')
plt.colorbar(im)

plt.show()
````

# Task 4 - Pearson Correlation and p-values

Use the `scipy.stats.pearsonr()` function to evaluate the Pearson Coefficient and the p-values for each parameter tested above. This will help you determine the parameters most likely to have a strong effect on the price of the laptops.


<!--## Change Log


````python
for var in ['Category', 'GPU', 'OS', 'CPU_core', 'Screen_Size_inch', 'CPU_frequency', 'RAM_GB', 'Storage_GB_SSD','Weight_pounds']:
    corr_coeffi, p_value = stats.pearsonr(df[var], df['Price'])
    print(var)
    print(f'Correlation Coeffient of {var} is {corr_coeffi} with P-value = {p_value}', '\n')
````

# Summarizing
