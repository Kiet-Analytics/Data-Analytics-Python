# Single Linear Regression

You have learnt that "CPU_frequency" is the parameter with the lowest p-value among the different features of the dataset. Create a single feature Linear Regression model that fits the pair of "CPU_frequency" and "Price" to find the model for prediction.


```python
# Write your code below and press Shift+Enter to execute 
lm = LinearRegression()
x = df[['CPU_frequency']]
y= df['Price']
lm.fit(x,y)

#predict training value
predicted_y = lm.predict(x)
```

Generate the Distribution plot for the predicted values and that of the actual values. How well did the model perform?


```python
# Write your code below and press Shift+Enter to execute 
ax1 = sns.distplot(y, hist=False, color='r', label='Actual Value')
sns.distplot(predicted_y, hist=False, color='b', label='Fitted Values', ax=ax1)
plt.xlabel('Price in dollars')
plt.ylabel('Proportion of laptops')
plt.legend()
plt.show()
```

Evaluate the Mean Squared Error and R^2 score values for the model. 


```python
squared_r_SLR = lm.score(x,y)
print(f'R^2 for this model is {squared_r_SLR}')

mse_SLR = mean_squared_error(y,predicted_y)
print(f'MSE for this model is {mse_SLR}')
```

#  Multiple Linear Regression
The parameters which have a low enough p-value so as to indicate strong relationship with the 'Price' value are 'CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU' and 'Category'. Use all these variables to create a Multiple Linear Regression system. 


```python
# Write your code below and press Shift+Enter to execute 
multi_lm = LinearRegression()
Z = df[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']]
multi_lm.fit(Z,y)

Yhat = multi_lm.predict(Z)
```

Plot the Distribution graph of the predicted values as well as the Actual values


```python
# Write your code below and press Shift+Enter to execute
ax1 = sns.distplot(y, hist=False, color="r", label="Actual Values")
sns.distplot(Yhat, hist=False, color="b", label="Fitted Values", ax=ax1)

plt.title('Actual vs Fitted Value')
plt.xlabel('Price in Dollars')
plt.ylabel('Proportion of Cars')
plt.legend()
plt.show()
```

Find the R^2 score and the MSE value for this fit. Is this better or worst than the performance of Single Linear Regression?


```python
# Write your code below and press Shift+Enter to execute
MLR_r2 = multi_lm.score(Z, y)
print(f"r^2 of this model is {MLR_r2}")

MLR_mse = mean_squared_error(y, Yhat)
print(f'MSE for this model is {MLR_mse}')
```

# Polynomial Regression
Use the variable "CPU_frequency" to create Polynomial features. Try this for 3 different values of polynomial degrees. Remember that polynomial fits are done using `numpy.polyfit`. 


```python
#degree1
x = x.flatten()
f1 = np.polyfit(x,y,1)
poly_model_1 = np.poly1d(f1)

# #degree2
f3 = np.polyfit(x,y,3)
poly_model_3 = np.poly1d(f3)

# #degree3
f5 = np.polyfit(x,y,5)
poly_model_5 = np.poly1d(f5)
```

Plot the regression output against the actual data points to note how the data fits in each case. To plot the polynomial response over the actual data points, you have the function shown below.


```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(independent_variable.min(),independent_variable.max(),100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title(f'Polynomial Fit for Price ~ {Name}')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of laptops')
```

```python
def Plotpolly(model,independent_variable, dependent_variable, Name):
    x_new = np.linspace(independent_variable.min(), independent_variable.max(), 100)
    y_new = model(x_new)

    plt.plot(independent_variable,dependent_variable, '.', x_new, y_new, '-' )
    plt.title(f'Polymonial Fit for Price ~ {Name}')

    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of laptops')
    
```

Call this function for the 3 models created and get the required graphs.


```python
# Call for function of degree 1
Plotpolly(poly_model_1, x, y, 'CPU_frequency')
```

```python
# Call for function of degree 3
Plotpolly(poly_model_3, x, y, 'CPU_frequency')
```

```python
#  Write your code below and press Shift+Enter to execute
# Call for function of degree 5
Plotpolly(poly_model_5, x, y, 'CPU_frequency')
```

Also, calculate the R^2 and MSE values for these fits. For polynomial functions, the function sklearn.metrics.r2_score will be used to calculate R^2 values.


```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#  Write your code below and press Shift+Enter to execute

poly1_r2_score = r2_score(y, poly_model_1(x))
print(f'The R-Square value for 1st degree polymonial is {poly1_r2_score}')
print(f'The MSE value for 1st degree polymonial is {mean_squared_error(y, poly_model_1(x))}','\n')


poly3_r2_score = r2_score(y, poly_model_3(x))
print(f'The R-Square value for 3rd degree polymonial is {poly3_r2_score}')
print(f'The MSE value for 3rd degree polymonial is {mean_squared_error(y, poly_model_3(x))}','\n')

poly5_r2_score = r2_score(y, poly_model_5(x))
print(f'The R-Square value for 5th degree polymonial is {poly5_r2_score}')
print(f'The MSE value for 5th degree polymonial is {mean_squared_error(y, poly_model_5(x))}','\n')

```

# Pipeline
Create a pipeline that performs parameter scaling, Polynomial Feature generation and Linear regression. Use the set of multiple features as before to create this pipeline. 


```python
#  Write your code below and press Shift+Enter to execute
Input = [
    ('scale', StandardScaler()),
    ('polynomial', PolynomialFeatures(include_bias=False)),
    ('model', LinearRegression())
]
pipe = Pipeline(Input)

Z = Z.astype(float)
pipe.fit(Z, y)
predicted_y_pipe = pipe.predict(Z)
```

Evaluate the MSE and R^2 values for the this predicted output.


```python
#  Write your code below and press Shift+Enter to execute
print(f"R^2 Score for multi-variable polynomial pipeline is {r2_score(y, predicted_y_pipe)}")
print(f"MSE  for multi-variable polynomial pipeline is {mean_squared_error(y, predicted_y_pipe)}")
```

I have seen that the values of R^2 increase as we go from Single Linear Regression to Multiple Linear Regression. Further, if we go for multiple linear regression extended with polynomial features, we get an even better R^2 value.
